```python
import pandas as pd
```

```python
from pymodulon.io import *

ica = load_json_model('../data/imods/precise1k_with_ygfZ_muts.json.gz')
```

```python
all_imod_act_df = ica.A.T
display(len(all_imod_act_df), all_imod_act_df.head())
```

```python
ygfZ_mut_char_exp_md_df = pd.read_csv('./30 samples (18 samples Melatonin inhibition re-extract + 12 samples ygfZ mutant).csv', index_col=0)
ygfZ_mut_char_exp_md_df = ygfZ_mut_char_exp_md_df[ygfZ_mut_char_exp_md_df.B.str.contains('SDT_ygfZ')].copy()
ygfZ_mut_char_exp_md_df.set_index('B', inplace=True)

def _get_description(d):
    s = ''
    if 'HMP3427' in d['D']:
        s += 'parent strain'
    if 'SDT495' in d['D']:
        s += 'YgfZ T108P mutant'
    if d['E'] == 'control':
        s += ' in control media'
    if d['E'] == '5 mM H2O2':
        s += ' in 5 mM H2O2 media'
    return s

ygfZ_mut_char_exp_md_df['Description'] = ygfZ_mut_char_exp_md_df.apply(lambda r: _get_description(r), axis=1)
ygfZ_mut_char_exp_md_df
```

```python
rename_sample_d = {}
for i, r in ygfZ_mut_char_exp_md_df.iterrows():
    rename_sample_d[i] = r['Description']

ygfZ_mut_char_imod_act_df = all_imod_act_df[all_imod_act_df.index.str.contains('SDT_ygfZ')].copy()
ygfZ_mut_char_imod_act_df.rename(index=rename_sample_d, inplace=True)
ygfZ_mut_char_imod_act_df.head()
```

```python
# Remove all columns with values only between 5 and -5
mask = (ygfZ_mut_char_imod_act_df.abs() <= 5).all()
ygfZ_mut_char_imod_act_df = ygfZ_mut_char_imod_act_df.loc[:, ~mask]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.dpi"] = 200

strain_clr_d = {
    'parent strain in control media': '#9EDAE5',
    'parent strain in 5 mM H2O2 media': '#17BECF',
    'YgfZ T108P mutant in control media': '#F7B6D2',
    'YgfZ T108P mutant in 5 mM H2O2 media': '#E377C2',
}
row_colors = pd.Series(strain_clr_d)

cm = sns.clustermap(
    ygfZ_mut_char_imod_act_df, cmap='coolwarm', cbar_kws={'label': 'iModulon activity'},
    figsize=(12.5, 3),
    # figsize=(30, 3),
    yticklabels=False,
    xticklabels=True,
    # cbar_kws={'label': 'distance (Ã…)', 'labelpos': 'left'},
    cbar_pos=(0.19, 0.45, 0.01, 0.4),
    center=0,  # For the cbar colors.
    row_colors=row_colors[ygfZ_mut_char_imod_act_df.index],
)
cm.cax.yaxis.set_label_position('left')

# move sample color to the right side
ax_row_colors = cm.ax_row_colors
box = ax_row_colors.get_position()
box_heatmap = cm.ax_heatmap.get_position()
ax_row_colors.set_position([box_heatmap.max[0]+0.005, box.y0, box.width*0.5, box.height])
ax_row_colors.set_xticks([])  # Remove the single xtick of the color row.

cm.ax_heatmap.set_xlabel('all iModulons')
cm.ax_heatmap.set_ylabel('sample', labelpad=22.5)
cm.ax_row_dendrogram.set_visible(False)
cm.ax_col_dendrogram.set_visible(False)
cm.ax_heatmap.tick_params(axis='both', which='both', length=0)
cm.ax_heatmap.set_xlabel('all iModulons with activities >= 5 or <= -5')

import matplotlib.patches as mpatches
patches = [mpatches.Patch(color=v, label=k) for k, v in strain_clr_d.items()]
plt.legend(title='$\it{E. coli}$ melatonin production strains', handles=patches, ncol=2, bbox_to_anchor=(40, 1.1), loc='lower center')

# Bold CP4-44 iModulon axis tick label.
label_to_bold = 'CP4-44'
xtl = [l.get_text() for l in cm.ax_heatmap.get_xticklabels()]
for i in range(len(xtl)):
    if xtl[i] == label_to_bold:
        cm.ax_heatmap.get_xticklabels()[i].set_weight('bold')

plt.savefig("f6c.svg")
```
